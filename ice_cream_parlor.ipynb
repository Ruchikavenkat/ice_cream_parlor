```python
#database managament class
class Databasemanagement:
    #connecting sqlite 
    def __init__(self,db_name="ice_cream_parlor.db"):
        self.connection=sqlite3.connect(db_name)
        self.cursor=self.connection.cursor()
        self.initialize_db()
    
    #initializing the database
    def initialize_db(self):
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS flavours(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        season TEXT NOT NULL)
        """)
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS ingredients(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL)
        """)
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS allergens(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL UNIQUE)
        """)
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS flavours(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        season TEXT NOT NULL)
        """)
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS customer_suggestions(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        flavour_name TEXT NOT NULL,
        allergens TEXT)
        """)
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS cart(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        flavour_name TEXT NOT NULL)
        """)
        self.connection.commit()
    
    #fetching and executing the required query     
    def execute_query(self,query,parameters=()):
        self.cursor.execute(query,parameters)
        self.connection.commit()
    def fetch_query(self,query,parameters=()):
        self.cursor.execute(query,parameters)
        return self.cursor.fetchall()
    def close_connection(self):
        self.connection.close()

   
```

```python
#ice_cream parlor class
class ice_cream_parlor:
    #reffering to the database class
    def __init__(self,db_management):
        self.db=db_management
    #adding values into the tables 
    def add_flavor(self,name,season):
        self.db.execute_query("INSERT INTO flavors(name,season)VALUES (?,?)",(name,season))
    def add_ingredient(self,name):
        self.db.execute_query("INSERT INTO ingredients(name) VALUES (?)",(name,))
    def add_allergen(self,name):
        try:
            self.db.execute_query("INSERT INTO allergens (name) VALUES (?)",(name,))
        except sqlite3.IntegrityError:
            print(f"Allergen {name} already exists")
    def add_suggestion(self,flovar_name,allergens):
        self.db.execute_query("INSERT INTO customer_suggestions(flavor_name,allergens)VALUES(?,?)",(flovar_name,allergens))
    def search_flavors(self,keyword,season=None):
        query="SELECT * FROM flavors WHERE name LIKE ?"
        parameters = [f"%{keyword}%"]
        if season:
            query += "AND season = ?"
            parameters.append(season)
        return self.db.fetch_query(query, parameters)
    def add_cart(self,flavor_name):
        self.db.execute_query("INSERT INTO cart(flavor) name VALUES(?)" , (flavor_name,))
    def view_cart(self):
        return self.db.fetch_query("SELECT * FROM cart")
```
